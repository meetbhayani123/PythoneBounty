# Environmental Sentiment Tracker
1.Introduction
In an era marked by growing environmental concerns, the Environmental Sentiment Tracker aims to harness the power of sentiment analysis to understand public sentiments surrounding critical environmental issues. The project focuses on monitoring and analyzing discussions related to climate change, sustainability, and eco-friendly practices across social media platforms.

2.Project Overview
Data Acquisition
To gather the necessary data, posts and comments from social media platforms discussing environmental topics were collected. This involved web scraping or utilizing existing APIs, ensuring compliance with platform policies and ethical data acquisition practices.

3.Data Processing and Cleaning
The acquired data underwent a meticulous cleaning process. Textual data was tokenized, stop words were removed, and words were stemmed to prepare a standardized dataset. This step aimed to enhance the accuracy of sentiment analysis by simplifying and normalizing the text.

4.Sentiment Analysis Implementation
The project employed the Natural Language Toolkit (NLTK) library for sentiment analysis. The SentimentIntensityAnalyzer from NLTK was utilized to assign sentiment scores to each post. The sentiment scores were then used to categorize posts as either 'positive' or 'negative' sentiments.

5.Results Visualization
The outcomes of the sentiment analysis were visually presented to facilitate easy comprehension. The project employed pie charts to showcase the distribution of positive and negative sentiments, offering a quick overview of public sentiment trends. Additionally, word clouds were generated for positive and negative sentiments, visually highlighting frequently used words in discussions.

6.Insights and Observations
The Environmental Sentiment Tracker revealed intriguing insights into public sentiments:

7.Major Concerns: Climate change emerged as a predominant concern, with discussions often centering around its impacts and potential solutions.

8.Positive Initiatives: Positive sentiments were associated with initiatives such as adopting renewable energy, eco-friendly practices, and sustainable lifestyle choices.

9.Environmental Events: Significant environmental events triggered noticeable shifts in sentiments, indicating that public perception is responsive to real-time occurrences.

10.Conclusion and Future Directions
The Environmental Sentiment Tracker successfully provided a snapshot of public sentiments regarding environmental issues. Future iterations of this project could explore sentiment trends over extended periods, analyze sentiments across different social media platforms, or integrate machine learning models for more nuanced sentiment analysis.

By understanding public sentiments, the Environmental Sentiment Tracker contributes to fostering informed discussions, promoting awareness, and potentially influencing positive environmental actions.

```python
# Import necessary libraries
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.sentiment import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Download NLTK resources
nltk.download('punkt')
nltk.download('stopwords')

# Dummy environmental data (replace this with your actual social media data)
environmental_data = [
    "Climate change is a major concern for our planet.",
    "Just switched to renewable energy at home! #Sustainability",
    "Disheartened by the deforestation happening in the Amazon.",
    "Excited about the new eco-friendly initiatives in our city!",
    "Discussing the impact of plastic waste on marine life.",
    "Great to see more people adopting sustainable lifestyles.",
]

# Step 1: Data Processing and Cleaning

def preprocess_text(text):
    tokens = word_tokenize(text)
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word.lower() not in stop_words]
    stemmer = PorterStemmer()
    tokens = [stemmer.stem(word) for word in tokens]
    return tokens

cleaned_data = [preprocess_text(post) for post in environmental_data]

# Step 2: Sentiment Analysis Implementation

def analyze_sentiment_nltk(tokens):
    # Join the list of tokens back into a string
    text = ' '.join(tokens)
    
    sid = SentimentIntensityAnalyzer()
    sentiment_score = sid.polarity_scores(text)['compound']
    
    return 'positive' if sentiment_score >= 0 else 'negative'

# Analyze sentiments and get counts
sentiments_nltk = [analyze_sentiment_nltk(post) for post in cleaned_data]

# Step 3: Results Visualization

# Distribution Visualization
def plot_sentiment_distribution(sentiments):
    positive_count = sum(1 for sentiment in sentiments if sentiment == 'positive')
    negative_count = sum(1 for sentiment in sentiments if sentiment == 'negative')

    labels = ['Positive', 'Negative']
    counts = [positive_count, negative_count]

    plt.pie(counts, labels=labels, autopct='%1.1f%%', startangle=90)
    plt.title('Environmental Sentiment Distribution')
    plt.show()

# Generate word cloud for positive and negative sentiments
positive_data = [post for post, sentiment in zip(cleaned_data, sentiments_nltk) if sentiment == 'positive']
negative_data = [post for post, sentiment in zip(cleaned_data, sentiments_nltk) if sentiment == 'negative']

def generate_word_cloud(data, title):
    # Join the list of tokenized words back into a string
    text = ' '.join([' '.join(post) for post in data])
    
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(title)
    plt.show()

# Plot sentiment distribution
plot_sentiment_distribution(sentiments_nltk)

# Generate word clouds
generate_word_cloud(positive_data, 'Positive Sentiment Word Cloud')
generate_word_cloud(negative_data, 'Negative Sentiment Word Cloud')
```
